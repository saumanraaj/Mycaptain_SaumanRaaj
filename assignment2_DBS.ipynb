```python
from google.colab import files
f=files.upload()
import json
import random
import numpy as np
import pandas as pd
with open('student_information.json','r')as f:
    student_data=json.load(f)
print(student_data)
#creating student data from the json file
n_students = student_data['number']
first_names = student_data['first_names']
last_names = student_data['last_names']
gpa_mean = 3.0
gpa_std = 0.5
gpa_min, gpa_max = student_data['GPA']['range']
status_distribution = student_data['status']
student_id_min = student_data['student_id']['min']
student_id_max = student_data['student_id']['max']
student_id_prefix = student_data['student_id']['prefix']
students=[]
for i in range(n_students):
  first_name = random.choice(first_names)
  last_name = random.choice(last_names)
  student_id = student_id_prefix + str(random.randint(student_id_min, student_id_max))
  gpa = max(gpa_min, min(gpa_max, np.random.normal(gpa_mean, gpa_std)))
  status = np.random.choice([status[1] for status in status_distribution], p=[status[0] for status in status_distribution])
  students.append({
        'First Name': first_name,
        'Last Name': last_name,
        'Student ID': student_id,
        'GPA': round(gpa, 2),
        'Status': status
  })

students_df=pd.DataFrame(students)
students_df.to_csv('student_database.csv',index=False)
print(students_df)
```
